```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
exp='Exp001';fre_exp='A01a'
yr=1991
cyr=str(yr)
path='../'+exp+'/Sections/data/Sections/Atl_26N_'+fre_exp+'_'+cyr+'.nc'
print(path)
ds=xr.open_dataset(path)
```

```python
ds
```

```python
lons=ds['Longitude'].load().data
zax=-ds['z_l'].load().data
zplt=ds['atl_vh'].mean(axis=0).load().data/1.e6
cf=plt.pcolormesh(lons,zax,np.squeeze(zplt),vmin=-.6,vmax=.6,cmap=plt.cm.bwr)
plt.colorbar(cf)
plt.xlim(-100,-85)
```

```python
lons=ds['Longitude'].load().data
zax=-ds['z_l'].load().data
zplt=ds['temp'].mean(axis=0).load().data
cf=plt.pcolormesh(lons,zax,np.squeeze(zplt),cmap=plt.cm.Spectral_r)
plt.colorbar(cf)
plt.xlim(-100,-85)
```
